```python
import pandas as pd
```

```python
df_inventory =  pd.read_csv("Inventory.csv")
df_inventory = df_inventory.drop_duplicates()

print(df_inventory.isna().sum(), '\n')
display(df_inventory[df_inventory.isna().any(axis=1)])

df_inventory = df_inventory.fillna('UNKNOWN')

df_inventory.to_csv('cleaned_Inventory.csv', index=False)
```

```python
df_outbound=  pd.read_csv("Outbound.csv")
df_outbound = df_outbound.dropna()

df_outbound.to_csv('cleaned_Outbound.csv', index=False)
```

```python
df_Inbound=  pd.read_csv("Inbound.csv")
df_Inbound = df_Inbound.dropna()

df_Inbound.to_csv('cleaned_Inbound.csv', index=False)
```

```python
df_Inbound['TYPE'] = 'Inbound'
df_outbound['TYPE'] = 'Outbound'
df_Inbound.rename(columns={'INBOUND_DATE': 'DATE'}, inplace=True)
df_outbound.rename(columns={'OUTBOUND_DATE': 'DATE'}, inplace=True)

movement_df = pd.concat([df_Inbound, df_outbound])
movement_df.to_csv("combined_movement.csv", index=False)
movement_df
```
